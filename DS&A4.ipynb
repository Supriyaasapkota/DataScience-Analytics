```python
#lab 4: Chi-squared Test of independence in python
#Step 1:Import required libraries
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
#step 2:Create the observed frequency  (contigency) table
#example:gender vs product preference
data = {
    "Like": [30, 20],
    "Dislike": [10, 40]
}
```

```python
observed= pd.DataFrame(data,index=["Male","Female"])
print("Observed Frequency Table:")
print(observed)
```

```python
#Step 3:Perform Chi-squared Test
chi2_stat,p_val,dof,expected=chi2_contingency(observed)
```

```python
# Step 4: Display results
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_val)

print("\nExpected Frequency Table:")
print(pd.DataFrame(expected, index=observed.index, columns=observed.columns))
```

```python
# Step 5: Hypothesis decision
alpha = 0.05

if p_val < alpha:
    print("\nReject the Null Hypothesis")
    print("--> There is a significant association between the variables.")
else:
    print("\nFail to reject the Null Hypothesis")
    print("--> There is no significant association between the variables.")
```
